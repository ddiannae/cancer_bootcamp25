```python
def clean_dataframe(df):
    threshold = len(df)* 0.5
    ## Drop columns with more than 50% NaN values
    df_cleaned = df.dropna(thresh=threshold, axis=1)
    ## Fill remaining NaN values with 0
    df_cleaned.fillna(0.0, inplace = True)
    return df_cleaned
```

## Getting correlations

Getting correlation matrix using corr

```python
correlation_matrix = combined_df.corr(method='pearson')
```

```python
correlation_matrix_sub = correlation_matrix.iloc[490:, 0:490]
correlation_matrix_sub
```

```python
abs_max = max(abs(correlation_matrix_sub.min().min()), abs(correlation_matrix_sub.max().max()))
g = sns.heatmap(correlation_matrix_sub, 
            xticklabels = False, 
            yticklabels = False, 
            cmap='seismic', 
            vmin = -abs_max,
            vmax = abs_max, 
            cbar_kws={"shrink": 0.5})

plt.xlabel('E3 ligase expression')
plt.ylabel('TF dependencies')
plt.show()
```

Getting correlation matrix using `pearsonr`

```python
pvals = []
corrs = []
tfs = []
e3s = []
for e3_col in E3_expr_filtered.columns:
    for tf_col in tfs_effect_filtered.columns:
        # Compute Pearson correlation and p-value
        corr, p_value = pearsonr(E3_expr_filtered[e3_col], tfs_effect_filtered[tf_col])
        # Append results to the DataFrame
        pvals.append(p_value)
        corrs.append(corr)
        tfs.append(tf_col)
        e3s.append(e3_col)

correlation_df = pd.DataFrame({
    "TF" : tfs,
    "E3_ligase" : e3s,
    "p-val": pvals,
    "corr": corrs
})
```

```python
correlation_df
```

```python
correlation_df = correlation_df.sort_values(by = "corr")
```

```python
correlation_df.head(n=15)
```

```python
correlation_df.to_pickle("pydata/tf_dependency_e3ls_expr_correlation.pkl")
```

```python
correlation_df.head()
```

```python
correlation_df['q-value'] = fdrcorrection(correlation_df['p-val'].values)[1]
```

```python
correlation_df.head()
```

```python
correlation_df["qlog10"] = np.log10(correlation_df["q-value"]) * -1
```

```python
correlation_df.head()
```

```python
correlation_df["name"] = correlation_df["TF"] + "-" + correlation_df["E3_ligase"]
```

```python
correlation_df.head()
```

```python
ax = plt.subplot(111)
sns.scatterplot(data = correlation_df[correlation_df["q-value"] >= 0.1], x = "corr", y = "qlog10",
                s = 1, color = "grey")
sns.scatterplot(data = correlation_df[correlation_df["q-value"] < 0.1], x = "corr", y = "qlog10",
                s = 3)
sns.scatterplot(data = correlation_df[correlation_df["qlog10"] > 100], x = "corr", y = "qlog10",
                s = 10, color = "red")
for i, row in correlation_df[correlation_df["qlog10"] > 100].iterrows():
    plt.text(x = row["corr"]+0.01, y = row["qlog10"], s = row["name"], fontsize = 8)
plt.title("TF-E3 ligases correlation")
plt.xlabel("Pearson correlation")
plt.ylabel("-log10(q-value)")
ax.spines[['right', 'top']].set_visible(False)

```

### Plotting lowest corr

```python
for i in range(10):
    tf = correlation_df.iloc[i]["TF"]
    e3_ligase = correlation_df.iloc[i]["E3_ligase"]
    df = pd.DataFrame({"tf" : tfs_effect_filtered[tf],
                       "e3" : E3_expr_filtered[e3_ligase],
                       "model_id": E3_expr_filtered.index.values})
    df = pd.merge(df, model[["ModelID", "OncotreeLineage"]], left_on="model_id", right_on="ModelID")
    m, b = np.polyfit(df["tf"].values, df["e3"].values, 1)
    ax = plt.subplot(111)
    sns.scatterplot(data = df, x = "tf", y = "e3", hue = "OncotreeLineage",
                     palette= df_palette)
    plt.ylim(0, max(E3_expr_filtered[e3_ligase]))
    ax.text(0, max(df["e3"])-0.25, f"r = {round(correlation_df.iloc[i]["corr"], 5)}", fontsize = 8)
    ax.text(0, max(df["e3"])-0.5, f"q-val = {correlation_df.iloc[i]["q-value"]:.3e}", fontsize = 8)
    ax.plot(df["tf"].values, m*df["tf"].values + b, color='grey', linewidth=1)
    ax.spines[['right', 'top']].set_visible(False)
    handles, labels = ax.get_legend_handles_labels()
    # Example: remove the first item
    new_handles = handles[0:15]
    new_labels = labels[0:15]
    #ax.legend(new_handles, new_labels, title="Category")
    plt.title(f"{tf} depedency - {e3_ligase} expression")
    plt.xlabel("TF effect score")
    plt.ylabel("E3 ligase expression")
    plt.legend(new_handles, new_labels, bbox_to_anchor = (1.5, 0.9), frameon=False)
    plt.show()
```
