```python
expr_data = tc.get(name='internal-24q4-8c04', version=101, file='OmicsExpressionProteinCodingGenesTPMLogp1BatchCorrected')
```

```python
expr_data.head()
```

```python
tfs = pd.read_csv("data/tf_list.csv")
tfs_reviewed = pd.read_excel("data/TF_names.xlsx")
tfs_reviewed = tfs_reviewed[["Original names", "new names"]]
tfs = tfs[~tfs["hugo_symbol"].isin(tfs_reviewed["Original names"].values)]
tfs["hugo_symbol"] = pd.concat([tfs["hugo_symbol"], tfs_reviewed["new names"]], ignore_index=True)
```

```python
expr_data.columns = [re.sub(r"\(\d+\)", "", c).strip() for c in expr_data.columns]
```

```python
expr_data.to_pickle("pydata/tpmlog2p1.pkl")
```

```python
tfs_found = [tf for tf in tfs["hugo_symbol"] if tf in expr_data.columns]
```

```python
tf_expr = expr_data.loc[:, tfs_found]
```
